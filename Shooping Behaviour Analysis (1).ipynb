##    Visualization

```python
bar=df.groupby("Location")["Purchase Amount (USD)"].sum().plot(kind='bar')
bar.set_title("Total Purchase Amount by Location")
bar.set_xlabel("Location")
bar.set_ylabel("Purchase Amount (USD)")
print(bar)
```

BarGraph:

-It show the total purchase amount by location 

-In x-axis we have locations and in y-axis we have purshase amount

```python
gender_ratings = df.groupby("Gender")["Review Rating"].count().plot(kind='bar',color='orange')
gender_ratings.set_title("Number of Review Ratings by Gender")
gender_ratings.set_xlabel("Gender")
gender_ratings.set_ylabel("Count of Reviews")
print(gender_ratings)
```

-It shows the number of review rating by gender

-In this  Males have more review ratings compared to females.

```python
ax=df.groupby(["Subscription Status", "Payment Method"])["Purchase Amount (USD)"].count().plot(kind='bar',color='pink')
ax.set_title("Purchase Count by Subscription Status and Payment Method")
ax.set_xlabel("Subscription Status, Payment Method")
ax.set_ylabel("Number of Purchases")
print(ax)
```

-The above bar shows the purshase Count by Subscription Status and Payment Method

```python
#df.groupby("Category")["Purchase Amount (USD)"].agg(["sum", "mean"]).plot(kind='bar')
df['Category'].value_counts().plot(kind='bar',title="Number of Items per Category",xlabel='Category',ylabel='Count',color='lightblue')
```

It shows the number of Items per Category

A bar chart with each category on the x-axis.

The count of rows in that category as bar height.

The number displayed above each bar.

```python

gender_sales = df.groupby("Gender")["Purchase Amount (USD)"].sum()

# Choose custom colors (you can add more if needed)
custom_colors =["#FF9999", "#66B2FF"]

plt.pie(gender_sales,labels=gender_sales.index,autopct='%1.1f%%',colors=custom_colors)
plt.title("Share of Purchases by Gender")
plt.show()
```

It shows gender-wise chart showing the percentage share of total purchase amounts, using pink for Female and blue for Male

```python
category_sales = df.groupby("Category")["Purchase Amount (USD)"].sum()
custom_colors = ["#FF9999", "#66B2FF", "#99FF99", "#FFD966"]
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%',colors=custom_colors)
plt.title("Share of Purchases by Category")
plt.show()
```

It shows a category-wise pie chart showing the share of purchase amounts across different product categories, with each category in a distinct custom color
In this we can see that clothing are more purchased that is 44.7% and outerwear are less purchased that is 7.9%


```python
plt.figure(figsize=(8,6))
sns.barplot(data=df, x="Category", y="Purchase Amount (USD)", palette="Set2",hue='Category',legend=False)
plt.title("Average Purchase by Category")
plt.xlabel("Category")
plt.ylabel("Average Purchase Amount (USD)")
plt.show()
```

```python
sns.histplot(df['Purchase Amount (USD)'], bins=12,color='olive')
plt.xlabel('Purchase Amount (USD)') 
plt.ylabel('Number of Customers')
plt.title('Distribution of Purchase Amounts')
plt.show()

```

It groups continuous data into ranges

```python
top_products = df.groupby("Item Purchased")["Purchase Amount (USD)"].sum().nlargest(10).reset_index()

plt.figure(figsize=(10,6))
plt.bar(top_products["Item Purchased"], top_products["Purchase Amount (USD)"],color='gold')
plt.title("Top 10 Products with Highest Total Purchase Amount")
plt.xlabel("Product")
plt.ylabel("Total Purchase Amount (USD)")
plt.xticks(rotation=45, ha="right")
plt.show()
```

Barplot:

-The 10 products that generated the most revenue.

-Bar = higher total purchase amount.

-Useful for identifying bestsellers.

```python
cor_values = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(cor_values, annot=True, fmt='.2f', cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Each cell shows how strongly two numeric columns are correlated.

1.00 = perfect positive correlation (moves together).

-1.00 = perfect negative correlation (moves oppositely).

0 = no correlation.

```python
plt.figure(figsize=(8,6))
sns.histplot(df['Review Rating'], bins=10, kde=True,color='brown')
plt.title("Review Ratings count")
plt.xlabel("Review Rating")
plt.ylabel("count")
plt.show()
```

Histplot:

-The distribution of review ratings (e.g., how many 1-star, 2-star, 5-star reviews).

-Helps you see if most customers give high ratings (positive feedback) or low ratings (negative feedback).

-The KDE curve smooths out the histogram to give a clearer shape of distribution.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="Category", y="Purchase Amount (USD)", data=df,color='teal')
plt.title("Boxplot: Purchase Amount by Category")
plt.show()
```

Boxplot:

How purchase amounts vary by category.

Which categories have higher or lower median purchases.

Which categories have outliers (very large purchases).

Spread of data (narrow vs wide box = low vs high variability).

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x="Age", y="Purchase Amount (USD)", hue="Gender", data=df)
plt.title("Scatter Plot: Age vs Purchase Amount")
plt.show()
```

1.Relationship between Age and Spending:

- If dots rise as age increases  older customers spend more.

- If flat  spending is independent of age.
  
- If scattered randomly  no clear relationship.ender Differences
  
2.Spread of Data (Variance)
  
- Wide vertical spread  big differences in spending within the same age group.
  
- Narrow spread  customers of similar age spend similarly.
  
3.Gender Differences:
  
-By comparing colors, you can see if one gender tends to purchase more at certain ages.

-Example: If female dots cluster higher  women spend more at that age group.

4.Outliers:

-Dots far away from the rest show unusually high or low purchases.

```python
pd.crosstab(df['Gender'], df['Subscription Status']).plot(kind="bar", stacked=True, figsize=(7,5))
plt.title("Stacked Bar Plot: Gender vs Subscription Status")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

Stacked bar plot:

-> For each gender, how many people fall into each subscription status category.

-> The height of the bar = total count for that gender.

-> The colored segments = breakdown by subscription status(orange-yes,blue-no)

```python
plt.figure(figsize=(8,5))
df.groupby("Age")["Purchase Amount (USD)"].mean().plot(kind="area", alpha=0.5,color='teal')
plt.title("Area Plot: Avg Purchase Amount by Age")
plt.xlabel("Age")
plt.ylabel("Avg Purchase Amount")
plt.show()
```

1.Area shows:

-> Shows how average spending changes with age.

-> The height of the filled area represents the average purchase amount for each age.

2.We can spot trends:

Rising  certain ages spend more.

Falling  certain ages spend less.

Flat  consistent spending across ages.

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Payment Method", data=df, order=df["Payment Method"].value_counts().index,color='grey')
plt.title("Count Plot: Payment Methods")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5)) 
df.groupby("Age")["Purchase Amount (USD)"].mean().plot(kind="density", alpha=0.5) 
plt.title("Area Plot: Avg Purchase Amount by Age")
plt.xlabel("Age") 
plt.ylabel("Avg Purchase Amount")
plt.show() 
```

This shows the Avg Purchase Amount by Age
The ages between 55-65 have more purchased amount

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x="Category", y="Purchase Amount (USD)", data=df,size=3)
plt.title("Swarmplot: Purchase Amount by Category")
plt.xticks(rotation=45)
plt.show()
```

This shows how purchase amounts are distributed across categories, with all points visible.

```python

# Find Top 10 items by total purchase amount
top_items = (
    df.groupby("Item Purchased")["Purchase Amount (USD)"]
    .sum()
    .nlargest(10)
    .index
)

# Filter the DataFrame for only those items
df_top_items = df[df["Item Purchased"].isin(top_items)]

# Plot violin plot
plt.figure(figsize=(12,6))
sns.violinplot(
    data=df_top_items,
    x="Item Purchased",
    y="Purchase Amount (USD)",
    palette="Set2"
)

plt.title("Distribution of Purchase Amount by Top 10 Items (Violin Plot)")
plt.xlabel("Item Purchased")
plt.ylabel("Purchase Amount (USD)")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
import plotly.express as px

# Aggregate purchase data
map_data = (
    df.groupby(["Location", "Season", "Item Purchased"])["Purchase Amount (USD)"]
    .sum()
    .reset_index()
)

# Create an interactive map (bubble map)
fig = px.scatter_geo(
    map_data,
    locations="Location",          # Country/City names
    locationmode="country names",  # Adjust if you have city names
    size="Purchase Amount (USD)",  # Bubble size = total purchases
    color="Item Purchased",        # Different colors for items
    hover_name="Item Purchased",   # Show item purchased on hover
    animation_frame="Season",      # Animate by season
    projection="natural earth"
)

fig.update_layout(
    title="Purchases by Item, Location, and Season",
    legend_title="Item Purchased"
)

fig.show()
```

Location  plotted on the map.

Size of bubble  purchase amount.

Color  item purchased.

Season  slider/animation to see changes across seasons.
