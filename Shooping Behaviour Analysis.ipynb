# SHOOPING BEHAVIOUR ANALYSIS
This notebook covers importing, cleaning, analyzing, and visualizing the Shooping_behavior_updated dataset using Pandas, NumPy, Matplotlib, and Seaborn. It includes 18 different plots.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

###    Visualization

```python
bar=df.groupby("Location")["Purchase Amount (USD)"].sum().plot(kind='bar')
bar.set_title("Total Purchase Amount by Location")
bar.set_xlabel("Location")
bar.set_ylabel("Purchase Amount (USD)")
print(bar)
```

```python
gender_ratings = df.groupby("Gender")["Review Rating"].count().plot(kind='bar')
gender_ratings.set_title("Number of Review Ratings by Gender")
gender_ratings.set_xlabel("Gender")
gender_ratings.set_ylabel("Count of Reviews")
print(gender_ratings)
```

```python
ax=df.groupby(["Subscription Status", "Payment Method"])["Purchase Amount (USD)"].count().plot(kind='bar')
ax.set_title("Purchase Count by Subscription Status and Payment Method")
ax.set_xlabel("Subscription Status, Payment Method")
ax.set_ylabel("Number of Purchases")
print(ax)
```

```python
#df.groupby("Category")["Purchase Amount (USD)"].agg(["sum", "mean"]).plot(kind='bar')
df['Category'].value_counts().plot(kind='bar',title="Number of Items per Category",xlabel='Category',ylabel='Count')
```

```python
gender_sales = df.groupby("Gender")["Purchase Amount (USD)"].sum()
plt.pie(gender_sales, labels=gender_sales.index, autopct='%1.1f%%')
plt.title("Share of Purchases by Gender")
plt.show()
```

```python
category_sales = df.groupby("Category")["Purchase Amount (USD)"].sum()
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%')
plt.title("Share of Purchases by Category")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=df, x="Category", y="Purchase Amount (USD)", palette="Set2",hue='Category',legend=False)
plt.title("Average Purchase by Category")
plt.xlabel("Category")
plt.ylabel("Average Purchase Amount (USD)")
plt.show()
```

```python
sns.histplot(df['Purchase Amount (USD)'], bins=12)
plt.xlabel('Purchase Amount (USD)') 
plt.ylabel('Number of Customers')
plt.title('Distribution of Purchase Amounts')
plt.show()

```

```python
top_products = df.groupby("Item Purchased")["Purchase Amount (USD)"].sum().nlargest(10).reset_index()

plt.figure(figsize=(10,6))
plt.bar(top_products["Item Purchased"], top_products["Purchase Amount (USD)"])
plt.title("Top 10 Products with Highest Total Purchase Amount")
plt.xlabel("Product")
plt.ylabel("Total Purchase Amount (USD)")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
cor_values = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(cor_values, annot=True, fmt='.2f', cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['Review Rating'], bins=10, kde=True,color='brown')
plt.title("Review Ratings count")
plt.xlabel("Review Rating")
plt.ylabel("count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="Category", y="Purchase Amount (USD)", data=df)
plt.title("Boxplot: Purchase Amount by Category")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x="Age", y="Purchase Amount (USD)", hue="Gender", data=df)
plt.title("Scatter Plot: Age vs Purchase Amount")
plt.show()
```

```python
pd.crosstab(df['Gender'], df['Subscription Status']).plot(kind="bar", stacked=True, figsize=(7,5))
plt.title("Stacked Bar Plot: Gender vs Subscription Status")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df.groupby("Age")["Purchase Amount (USD)"].mean().plot(kind="area", alpha=0.5)
plt.title("Area Plot: Avg Purchase Amount by Age")
plt.xlabel("Age")
plt.ylabel("Avg Purchase Amount")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Payment Method", data=df, order=df["Payment Method"].value_counts().index)
plt.title("Count Plot: Payment Methods")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.regplot(x="Age", y="Purchase Amount (USD)", data=df, scatter_kws={"alpha":0.5})
plt.title("Regplot: Age vs Purchase Amount")
plt.show()
```

```python
# Lineplot: Age vs Avg Purchase Amount
plt.figure(figsize=(8,5))
sns.lineplot(x="Age", y="Purchase Amount (USD)", data=df, estimator="mean")
plt.title("Lineplot: Avg Purchase Amount by Age")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x="Category", y="Purchase Amount (USD)", data=df,size=3)
plt.title("Swarmplot: Purchase Amount by Category")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
df.groupby("Age")["Purchase Amount (USD)"].mean().plot(kind="density", alpha=0.5)
plt.title("Area Plot: Avg Purchase Amount by Age")
plt.xlabel("Age")
plt.ylabel("Avg Purchase Amount")
plt.show()
```
